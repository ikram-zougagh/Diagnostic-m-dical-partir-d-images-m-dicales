```python
import os
import cv2

# Chemin du répertoire contenant votre dataset
path_to_dataset = "/content/drive/MyDrive/PFA/cleaned_dataset/Brain_tumor/Training"

# Chemin du répertoire où vous souhaitez stocker les images prétraitées
preprocessed_path = "/content/drive/MyDrive/PFA/pre_dataset/training"

# Créer le répertoire s'il n'existe pas
os.makedirs(preprocessed_path, exist_ok=True)

# Fonctions de prétraitement
def resize_images(images, size=(224, 224)):
    resized_images = []
    for img in images:
        resized_img = cv2.resize(img, size)
        resized_images.append(resized_img)
    return resized_images

def normalize_images(images):
    normalized_images = []
    for img in images:
        normalized_img = img / 255.0
        normalized_images.append(normalized_img)
    return normalized_images

def sharping(normalized_img):
    laplacian = cv2.Laplacian(normalized_img, cv2.CV_64F)
    sharp_img = normalized_img - 0.7 * laplacian
    return sharp_img

# Parcourir les sous-répertoires dans le dataset
for subdirectory in sorted(os.listdir(path_to_dataset)):
    subdirectory_path = os.path.join(path_to_dataset, subdirectory)

    # Créer un sous-répertoire correspondant dans l'emplacement prétraité
    output_subdirectory_path = os.path.join(preprocessed_path, subdirectory)
    os.makedirs(output_subdirectory_path, exist_ok=True)

    # Parcourir les fichiers images dans le sous-répertoire
    for filename in sorted(os.listdir(subdirectory_path)):
        img_path = os.path.join(subdirectory_path, filename)
        img = cv2.imread(img_path)

        # Vérifier si le fichier est une image valide
        if img is not None:
            # Appliquer les étapes de prétraitement
            resized_img = resize_images([img], size=(224, 224))[0]
            normalized_img = normalize_images([resized_img])[0]
            sharpened_img = sharping(normalized_img)

            # Enregistrer l'image prétraitée dans l'emplacement prétraité
            output_img_path = os.path.join(output_subdirectory_path, filename)
            cv2.imwrite(output_img_path, sharpened_img * 255)

            print(f"Prétraitement terminé pour {filename} dans {subdirectory}")

print("Prétraitement terminé sur l'ensemble du dataset.")
```

```python
import os
import cv2

# Chemin du répertoire contenant votre dataset
path_to_dataset = "/content/drive/MyDrive/PFA/cleaned_dataset/Brain_tumor/Testing"

# Chemin du répertoire où vous souhaitez stocker les images prétraitées
preprocessed_path = "/content/drive/MyDrive/PFA/pre_dataset/Testing"

# Créer le répertoire s'il n'existe pas
os.makedirs(preprocessed_path, exist_ok=True)

# Fonctions de prétraitement
def resize_images(images, size=(224, 224)):
    resized_images = []
    for img in images:
        resized_img = cv2.resize(img, size)
        resized_images.append(resized_img)
    return resized_images

def normalize_images(images):
    normalized_images = []
    for img in images:
        normalized_img = img / 255.0    # puisque  normalisation  suit le calcul suivant: (pixel_origine-min_pixels)/(max_pixels-min_pixels) sachant que le min est 0 et le max est 255
        normalized_images.append(normalized_img)
    return normalized_images

def sharping(normalized_img):
    laplacian = cv2.Laplacian(normalized_img, cv2.CV_64F)
    sharp_img = normalized_img - 0.7 * laplacian
    return sharp_img

# Parcourir les sous-répertoires dans le dataset
for subdirectory in sorted(os.listdir(path_to_dataset)):
    subdirectory_path = os.path.join(path_to_dataset, subdirectory)

    # Créer un sous-répertoire correspondant dans l'emplacement prétraité
    output_subdirectory_path = os.path.join(preprocessed_path, subdirectory)
    os.makedirs(output_subdirectory_path, exist_ok=True)

    # Parcourir les fichiers images dans le sous-répertoire
    for filename in sorted(os.listdir(subdirectory_path)):
        img_path = os.path.join(subdirectory_path, filename)
        img = cv2.imread(img_path)

        # Vérifier si le fichier est une image valide
        if img is not None:
            # Appliquer les étapes de prétraitement
            resized_img = resize_images([img], size=(224, 224))[0]
            normalized_img = normalize_images([resized_img])[0]
            sharpened_img = sharping(normalized_img)

            # Enregistrer l'image prétraitée dans l'emplacement prétraité
            output_img_path = os.path.join(output_subdirectory_path, filename)
            cv2.imwrite(output_img_path, sharpened_img * 255)

            print(f"Prétraitement terminé pour {filename} dans {subdirectory}")

print("Prétraitement terminé sur l'ensemble du dataset.")
```
