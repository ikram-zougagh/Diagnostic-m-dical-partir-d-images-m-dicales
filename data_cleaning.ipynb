```python
import os
import hashlib
import shutil

def calculate_file_hash(file_path):
    # Open the file and calculate its hash
    with open(file_path, "rb") as f:
        file_hash = hashlib.md5(f.read()).hexdigest()
    return file_hash

def deduplicate_images(dataset_path, output_path):
    # Dictionary to store file hashes
    hash_dict = {}

    # Walk through each file in the dataset directory
    for root, _, files in os.walk(dataset_path):
        for file in files:
            # Full path of the image file
            image_path = os.path.join(root, file)

            # Calculate the hash of the image file
            file_hash = calculate_file_hash(image_path)

            # Check if the hash already exists in the dictionary
            if file_hash in hash_dict:
                print(f"Duplicate found: {image_path}")
                # Move the duplicate to the output directory
                duplicate_output_dir = os.path.join(output_path, "duplicates")
                os.makedirs(duplicate_output_dir, exist_ok=True)
                shutil.move(image_path, duplicate_output_dir)
            else:
                # Add the hash to the dictionary
                hash_dict[file_hash] = image_path

    print("Deduplication completed.")
```

```python
# Chemin vers le répertoire contenant les images classées par catégorie
dataset_path = "/content/drive/MyDrive/cleaned_dataset/Brain_tumor/Training"

# Chemin vers le répertoire de sortie pour les images dédupliquées
output_path = "/content/drive/MyDrive/cleaned_dataset/Training"


# Appeler la fonction de déduplication et stockage
deduplicate_images(dataset_path, output_path)
```
